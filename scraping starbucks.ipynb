```python
#import requests to BeautifulSoup
from bs4 import BeautifulSoup 
import requests 
#import pandas for dataframe
import pandas as pd 
#url to Starbuck's Nutition Facts 
url = 'https://www.fastfoodmenuprices.com/starbucks-nutrition/'
```

```python
#function to gather nutrition facts for bottled drinks 
def nutritionfacts():
  #creating empty lists for names, calories, carbs, sodium 
  names = []
  calories = []
  carbs = []
  sodium = []

  #requesting data from url 
  page = requests.get(url)
  #page.status_code
  soup = BeautifulSoup(page.content, 'html.parser')

  #extracting the raw names of the drinks in column 1 
  rawnames = soup.find_all('td', class_ = 'column-1')
  #extracting the raw calories of the drinks in column 3
  rawcalories = soup.find_all('td', class_ = 'column-3')
  #extracting the raw carbs of the drinks in column 5
  rawcarbs = soup.find_all('td', class_ = 'column-5')
  #extracting the raw sodium of the drinks in column 8
  rawsodium = soup.find_all('td', class_ = 'column-8')


  #creating a for loop to get the names,calories, carbs, and sodium of bottled drinks
    #rows 165:215 for names and 160:210 for calories, carbs, and sodium
  #using .append function to add data into the values for the empty lists 
  for name in rawnames[165:215]:   
    names.append(name.get_text())
  for calorie in rawcalories[160:210]:
    calories.append(calorie.get_text())
  for carb in rawcarbs[160:210]:
    carbs.append(carb.get_text())
  for sod in rawsodium[160:210]:
    sodium.append(sod.get_text())

  #creating a csv file
  filename = "Starbuck's Nutrition Facts.csv"
  #loading the csv file into a dataframe
  df = pd.DataFrame()
  #variable names 
  df['Names'] = names
  df['Calories'] = calories
  df['Carbs'] = carbs 
  df['Sodium'] = sodium 
  #converting the dataframe into a csv file 
  df.to_csv(filename, index=False)
  return df
```

```python
#dataframe of the nutrition facts
StarbucksNutritionFacts = nutritionfacts()
```
